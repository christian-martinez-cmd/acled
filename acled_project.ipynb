The plot above doesn't give us a good visual so we'll map out locations where fatalities occurred

```python
#Define coordinates of where we want to center our map (Centered in USA)
usa_coords = [40.015, -95.2705]

#Create the map
mmap = folium.Map(location = usa_coords, zoom_start = 4)

#Display the map
mmap
```

```python
for index, dfusa in dfusa.iterrows():
    location = [dfusa['latitude'], dfusa['longitude']]
    folium.Marker(location, popup = f'Location:{dfusa["location2"]}\n Fatalities:{dfusa["fatalities"]}\n Event:{dfusa["event_type"]}').add_to(mmap)

#for index, franchise in franchises.iterrows():
#location = [franchise['latitude'], franchise['longitude']]
#folium.Marker(location, popup = f'Name:{franchise["store"]}\n Revenue($):{franchise["revenue"]}').add_to(map_kenya)

mmap
```

Now that we have exact points, we'll create a heatmap

```python
#Define coordinates of where we want to center our map (Centered in USA)
usa_coords = [40.015, -95.2705]

#Create the map
my_map = folium.Map(location = usa_coords, zoom_start = 4)

#Display the map
my_map
```

```python
from folium.plugins import HeatMap
df_copy = dfusa[dfusa.fatalities>0].copy()
df_copy['count'] = 1
base_map = my_map
HeatMap(data=df_copy[['latitude', 'longitude', 'count']].groupby(['latitude', 'longitude']).sum().reset_index().values.tolist(), radius=8, max_zoom=13).add_to(base_map)
```

```python
url = ("https://raw.githubusercontent.com/python-visualization/folium/master/examples/data")
state_geo = f"{url}/us-states.json"
```

```python
# initialize the map and store it in a m object
m = folium.Map(location=[40, -95], zoom_start=4)

# show the map
m
```

```python
m.choropleth(
    geo_data=state_geo,
    name="ACLED",
    data=dfusa,
    columns=["state", "fatalities"],
    key_on="properties.name",
    fill_color="YlGnBu",
    fill_opacity=1,
    fill=True,
    line_opacity=0.2,
    smooth_factor=0,
    Highlight=True,
    show=False,
    overlay=True,
    legend_name="Fatalities per state")

folium.LayerControl().add_to(m)

m

```
